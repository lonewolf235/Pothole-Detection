```python
import cv2
import os
import numpy as np
from PIL import Image

def sharpen_image(image_path):
    image = cv2.imread(image_path)
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened_image = cv2.filter2D(image, -1, kernel)

    return sharpened_image

data_dir="D:/Datasets/pothole_dataset/Pothole main dataset/train/images"
for filename in os.listdir(data_dir):
    # Read the image
    img = PIL.Image.open(os.path.join(data_dir, filename))
    # Apply the transformations to the image
    augmented_img = sharpen_image(img)

    # Save the augmented image
    augmented_img.save(os.path.join(data_dir, "sharpen_" + filename))
```

```python
#Resizer code

import os
from PIL import Image

# Get the path to the folder containing the images
image_folder = "D:/Datasets/pothole_dataset/Pothole main dataset/train/images"

# Get the width and height of the resized images
width = 640
height = 640

# Iterate over the images in the folder
for image in os.listdir(image_folder):

    # Open the image
    img = Image.open(os.path.join(image_folder, image))

    # Resize the image
    img = img.resize((width, height))

    # Save the resized image
    img.save(os.path.join(image_folder, image))
```

```python
#Image augmentation code

import torchvision.transforms as transforms

# Set the directory of the images to be augmented
data_dir = image_folder

# Define the transformations to be applied
transform = transforms.Compose(
    [
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    ]
)
# Iterate over the images in the directory
for filename in os.listdir(data_dir):
    # Read the image
    img = PIL.Image.open(os.path.join(data_dir, filename))

    # Apply the transformations to the image
    augmented_img = transform(img)

    # Save the augmented image
    augmented_img.save(os.path.join(data_dir, "augmented_" + filename))
```

```python
model = YOLO('D:\Datasets\pothole_dataset/runs\detect/train8\weights/best.pt')
image_path="D:\pothole.webp"
results=model(image_path)
```

```python
import cv2
img=cv2.imread(image_path)
for result in results:
    boxes = result.boxes.cpu().numpy() # get boxes on cpu in numpy
    for box in boxes: # iterate boxes
        c=box.cls
        name=model.names[int(c)]
        conf=(str)(round(box.conf[0],2))
        name=name+" "+conf
        print((name))
        r = box.xyxy[0].astype(int) # get corner points as int
        print(r) # print boxes
        cv2.rectangle(img, r[:2], r[2:], (0,255,0), 2) # draw boxes on img
        cv2.putText(
        img,
        name,
        (int(r[0]), int(r[1]) - 10),
        fontFace = cv2.FONT_HERSHEY_SIMPLEX,
        fontScale = 0.3,
        color = (0, 255,255),
        thickness=1
    )
        cv2.imwrite("output2.jpg", img)
```
